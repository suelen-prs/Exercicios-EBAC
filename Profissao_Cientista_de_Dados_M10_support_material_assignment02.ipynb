Considere a mesma base de dados utilizada na tarefa 1 ```previsao_de_renda.csv```.

1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
````

````python
url = 'https://raw.githubusercontent.com/dsvitorchaves/EBAC/main/previsao_de_renda.csv'
df = pd.read_csv(url)
df
````

## Insight 1:
Usando o boxplot, não conseguimos ter uma visualização muito exata, aparentemente tanto faz para o aumento da renda possuir veiculo ou imóvel.

````python
sns.boxplot(x='posse_de_veiculo', y='renda', data=df)
plt.title('Boxplot da Renda por Posse de Veículo')
plt.show()
````

````python
sns.boxplot(x='posse_de_imovel', y='renda', data=df)
plt.title('Boxplot da Renda por Posse de Imóvel')
plt.show()
````

## Insight 2:
Atraves das médias visualizadas em gráfico de barras, podemos ver uma diferença no aumento da renda quando existe a posse de veiculo e a posse de imóvel.

````python
media_veiculo = df.groupby('posse_de_veiculo')['renda'].mean()
media_imovel = df.groupby('posse_de_imovel')['renda'].mean()
````

````python
# Gráfico de barras - médias de 'renda' por 'posse_de_veiculo'
media_veiculo.plot(kind='bar')
plt.ylabel('Renda Média')
plt.title('Renda Média por Posse de Veículo')
plt.show()
````

````python
# Gráfico de barras - médias de 'renda' por 'posse_de_imovel'
media_imovel.plot(kind='bar')
plt.ylabel('Renda Média')
plt.title('Renda Média por Posse de Imóvel')
plt.show()
````

## Insight 3
Finalmente conseguimos visualizar atravez do pointplot, uma diferença bem consideravel na renda entre quem possui veiculo e/ou imóvel.

````python
# Pointplot - 'posse_de_veiculo' e 'renda'
plt.figure(figsize=(8, 5))
sns.pointplot(x='posse_de_veiculo', y='renda', data=df, capsize=.1)
plt.title('Renda Média por Posse de Veículo')
plt.ylabel('Renda Média')
plt.xlabel('Posse de Veículo')
plt.show()
````

````python
# Poinplot 'posse_de_imovel' e 'renda'
plt.figure(figsize=(8, 5))
sns.pointplot(x='posse_de_imovel', y='renda', data=df, capsize=.1)
plt.title('Renda Média por Posse de Imóvel')
plt.ylabel('Renda Média')
plt.xlabel('Posse de Imóvel')
plt.show()
````
