#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esse arquivo.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sinasc = pd.read_csv('b1abb33c4f.csv')
sinasc.head()
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
sinasc.loc[sinasc['APGAR5']>= 8, 'TPASFIXIA'] = 'normal'
sinasc.loc[(sinasc['APGAR5'] >= 6) & (sinasc['APGAR5']<=7), 'TPASFIXIA'] = 'asfixia leve'
sinasc.loc[(sinasc['APGAR5'] >= 4) & (sinasc['APGAR5']<=5), 'TPASFIXIA'] = 'asfixia moderada'
sinasc.loc[sinasc['APGAR5']<=3, 'TPASFIXIA'] = 'asfixia severa'
sinasc[['APGAR5']].dropna()
sinasc.head()
```

```python
sns.countplot(x= 'TPASFIXIA'
              , hue = 'TPASFIXIA'
              , data= sinasc
             )
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
sns.scatterplot(x = 'APGAR5'
                , y = 'SEXO'
                , data = sinasc
                , hue = 'GESTACAO'
                )
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tabela = pd.crosstab(sinasc['GESTACAO'], sinasc['APGAR5'])
tabela
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
percentual_linha =tabela.div(tabela.sum(axis=1),axis=0)
percentual_linha
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
percentual_linha.plot.bar()
```

Podemos concluir que o APGAR5 entre 32 a 42 semanas está em 9.0. Ou seja, o risco de uma criança morrer de asfixia é menor do que nas semanas anteriores a 32. A partir de 37 semanas, o parto poderia ser realizado, não categorizando prematuro.
