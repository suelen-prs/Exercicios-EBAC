### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
peso_medio = data_filtrada.groupby(['SEXO', 'ESCMAE'])['PESO'].mean()
print(peso_medio)
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
# Encontrar o município com menos bebês nascidos em 2019
municipio_com_menos_bebes = sinasc_raw['munResNome'].value_counts().idxmin()
municipio_com_menos_bebes
```

```python
# Filtrar o dataset para apenas esse município
data_municipio = sinasc_raw[sinasc_raw['munResNome'] == municipio_com_menos_bebes]
data_municipio
```

```python
# Calcular as estatísticas para as mães
idade_media_mae = data_municipio['IDADEMAE'].mean()
idade_maxima_mae = data_municipio['IDADEMAE'].max()
idade_minima_mae = data_municipio['IDADEMAE'].min()

# Calcular as estatísticas para os pais
idade_media_pai = data_municipio['IDADEPAI'].mean()
idade_maxima_pai = data_municipio['IDADEPAI'].max()
idade_minima_pai = data_municipio['IDADEPAI'].min()

print(f"Município com menos nascimentos em 2019: {municipio_com_menos_bebes}")
print(f"Idade média das mães: {idade_media_mae:.2f}")
print(f"Idade máxima das mães: {idade_maxima_mae}")
print(f"Idade mínima das mães: {idade_minima_mae}")
print(f"Idade média dos pais: {idade_media_pai:.2f}")
print(f"Idade máxima dos pais: {idade_maxima_pai}")
print(f"Idade mínima dos pais: {idade_minima_pai}")
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
data_marco = sinasc_raw[sinasc_raw['DTNASC'].str.contains('-03-')]
data_marco
```

```python
municipio_com_mais_bebes = data_marco['munResNome'].value_counts().idxmax()
municipio_com_mais_bebes
```

```python
data_municipio = data_marco[data_marco['munResNome'] == municipio_com_mais_bebes]
data_municipio
```

```python
qtd_media_filhos = data_municipio['QTDFILVIVO'].mean()
qtd_maxima_filhos = data_municipio['QTDFILVIVO'].max()
qtd_minima_filhos = data_municipio['QTDFILVIVO'].min()
print(f"Quantidade média de filhos vivos: {qtd_media_filhos:.2f}")
print(f"Quantidade máxima de filhos vivos: {qtd_maxima_filhos}")
print(f"Quantidade mínima de filhos vivos: {qtd_minima_filhos}")
```

```python
idade_media_pai = data_municipio['IDADEPAI'].mean()
idade_maxima_pai = data_municipio['IDADEPAI'].max()
idade_minima_pai = data_municipio['IDADEPAI'].min()
print(f"Idade média dos pais: {idade_media_pai:.2f}")
print(f"Idade máxima dos pais: {idade_maxima_pai}")
print(f"Idade mínima dos pais: {idade_minima_pai}")
```
